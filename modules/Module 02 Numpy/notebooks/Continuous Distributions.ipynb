```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, uniform, lognorm, expon

# Set a seed for reproducibility
np.random.seed(42)
```

# Continuous Probability Distributions
A continuous distribution describes the probabilities of a continuous random variable's possible values. A continuous random variable has an infinite and uncountable set of possible values (known as the range). The mapping of time can be considered as an example of the continuous probability distribution. It can be from 1 second to 1 billion seconds, and so on.

The area under the curve of a continuous random variable's PDF is used to calculate its probability. As a result, only value ranges can have a non-zero probability. A continuous random variable's probability of equaling some value is always zero.



## Normal Distribution
Normal Distribution is one of the most basic continuous distribution types. Gaussian distribution is another name for it. Around its mean value, this probability distribution is symmetrical. It also demonstrates that data close to the mean occurs more frequently than data far from it. Here, the mean is 0, and the variance is a finite value.



```python
# Example 3: Normal Distribution
mean = 0  # Mean of the normal distribution
std_dev = 1  # Standard deviation of the normal distribution

# Generate random samples from a normal distribution
normal_samples = np.random.normal(mean, std_dev, size=1000)
normal_samples[1:50]
```

```python
# Plot the histogram of normal samples
plt.hist(normal_samples, bins=30, density=True, alpha=0.75)
plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.show()
```

##  Uniform Distribution

```python
# Define the range of the uniform distribution
low = 0
high = 1

# Generate random samples from a uniform distribution
uniform_samples = uniform.rvs(loc=low, scale=high-low, size=1000)

uniform_samples[1:50]
```

```python
# Plot the histogram of uniform samples
plt.hist(uniform_samples, bins=30, density=True, alpha=0.75)
plt.title('Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.show()
```

## Log-Normal Distribution


```python
# Parameters of the log-normal distribution
mu = 0  # Mean of the natural logarithm of the distribution
sigma = 0.1  # Standard deviation of the natural logarithm of the distribution

# Generate random samples from a log-normal distribution
lognormal_samples = lognorm.rvs(s=sigma, scale=np.exp(mu), size=1000)
lognormal_samples[1:50]
```

```python
# Plot the histogram of log-normal samples
plt.hist(lognormal_samples, bins=30, density=True, alpha=0.75)
plt.title('Log-Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.show()
```

## Exponential Distribution

# Generate random samples from an exponential distribution

```python
# Rate parameter (lambda) of the exponential distribution
lambda_param = 0.5

# Generate random samples from an exponential distribution
exponential_samples = expon.rvs(scale=1/lambda_param, size=1000)
exponential_samples[1:50]
```

```python
# Plot the histogram of exponential samples
plt.hist(exponential_samples, bins=30, density=True, alpha=0.75)
plt.title('Exponential Distribution')
plt.xlabel('Time Until Next Event')
plt.ylabel('Probability Density')
plt.show()
```
