```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, poisson, norm, bernoulli

# Set a seed for reproducibility
np.random.seed(42)
```

# Discrete Probability Distribution
- Deals with discrete random variables, where the possible outcomes are countable.
- Example: The probability distribution of rolling a six-sided die.

## Binomial Distribution
The binomial distribution is a discrete distribution with a finite number of possibilities. When observing a series of what are known as Bernoulli trials, the binomial distribution emerges. A Bernoulli trial is a scientific experiment with only two outcomes: success or failure.

```python
# Binomial Distribution
n_trials = 10  # Number of trials
p_success = 0.5  # Probability of success in each trial

# Generate random samples from a binomial distribution
binomial_samples = np.random.binomial(n_trials, p_success, size=1000)
binomial_samples[1:10]
```

```python
# Plot the histogram of binomial samples
plt.hist(binomial_samples, bins=np.arange(0, n_trials + 2) - 0.5, density=True, alpha=0.75)
plt.title('Binomial Distribution')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.show()
```

## Bernoulli's Distribution
The Bernoulli distribution is a variant of the Binomial distribution in which only one experiment is conducted, resulting in a single observation. As a result, the Bernoulli distribution describes events that have exactly two outcomes.



```python
# Probability of success
p_success = 0.3

# Generate random samples from a Bernoulli distribution
bernoulli_samples = bernoulli.rvs(p_success, size=1000)
bernoulli_samples[1:50]
```

```python
# Plot the histogram of Bernoulli samples
plt.hist(bernoulli_samples, bins=[-0.5, 0.5, 1.5], density=True, alpha=0.75)
plt.xticks([0, 1], ['Failure (0)', 'Success (1)'])
plt.title('Bernoulli Distribution')
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.show()
```

## Poisson Distribution
A Poisson distribution is a probability distribution used in statistics to show how many times an event is likely to happen over a given period of time. To put it another way, it's a count distribution. Poisson distributions are frequently used to comprehend independent events at a constant rate over a given time interval. Siméon Denis Poisson, a French mathematician, was the inspiration for the name.



```python
# Poisson Distribution
lambda_param = 3  # Average number of events in a fixed interval

# Generate random samples from a Poisson distribution
poisson_samples = np.random.poisson(lambda_param, size=1000)
poisson_samples[1:20]
```

```python
# Plot the histogram of Poisson samples
plt.hist(poisson_samples, bins=np.arange(0, 2 * lambda_param + 2) - 0.5, density=True, alpha=0.75)
plt.title('Poisson Distribution')
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.show()
```
